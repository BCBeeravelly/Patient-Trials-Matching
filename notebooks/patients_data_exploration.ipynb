```python
import xml.etree.ElementTree as ET
import os
import pandas as pd

xml_path = '/Users/bharathbeeravelly/Desktop/patient-trials-matching/data/xml-exploration-2.xml'

```

```python
ns = {'hl7': 'urn:hl7-org:v3'}
```

```python
# Parse the XML file
tree = ET.parse(xml_path)
root = tree.getroot()
```

```python
# Find all 'id' elements
for id_element in root.findall('hl7:id', ns):
    # Extract the 'extension' attribute
    extension_value = id_element.get('extension')
    if extension_value:
        print(f"Extension: {extension_value}")
```

```python
# Find the recordTarget element and navigate inside it
record_target = root.find('hl7:recordTarget/hl7:patientRole/hl7:patient', ns)

if record_target is not None:
    # Extract administrativeGenderCode
    gender_code = record_target.find('hl7:administrativeGenderCode', ns)
    if gender_code is not None:
        gender = gender_code.get('code')
        print(f"Gender Code: {gender}")
    
    # Extract birthTime
    birth_time = record_target.find('hl7:birthTime', ns)
    if birth_time is not None:
        birth_value = birth_time.get('value')
        print(f"Birth Time: {birth_value}")
    
    # Extract raceCode displayName
    race_code = record_target.find('hl7:raceCode', ns)
    if race_code is not None:
        race_display_name = race_code.get('displayName')
        print(f"Race: {race_display_name}")
    
    # Extract ethnicGroupCode displayName
    ethnic_group = record_target.find('hl7:ethnicGroupCode', ns)
    if ethnic_group is not None:
        ethnic_display_name = ethnic_group.get('displayName')
        print(f"Ethnic Group: {ethnic_display_name}")
    
    # Extract languageCode code
    language_code = record_target.find('hl7:languageCommunication/hl7:languageCode', ns)
    if language_code is not None:
        language = language_code.get('code')
        print(f"Language Code: {language}")
```

```python
# Navigate to the structuredBody and then to the component sections
structured_body = root.find('.//hl7:structuredBody', ns)

# Now loop through all components within the structuredBody
for component in structured_body.findall('hl7:component', ns):
    section = component.find('hl7:section', ns)
    if section is not None:
        title = section.find('hl7:title', ns)
        if title is not None and 'Allergies' in title.text:
            print(f"Section: {title.text}")
            
            # Extract table rows under the text element
            rows = section.findall('.//hl7:tbody/hl7:tr', ns)
            for row in rows:
                start = row.find('hl7:td[1]', ns).text  # Start date
                stop = row.find('hl7:td[2]', ns).text  # Stop date (if present)
                description = row.find('hl7:td[3]', ns).text  # Allergy description
                code = row.find('hl7:td[4]', ns).text  # Allergy code

                print(f"Allergy Info:\n  Start: {start}\n  Stop: {stop if stop else 'N/A'}\n  Description: {description}\n  Code: {code}")
```
